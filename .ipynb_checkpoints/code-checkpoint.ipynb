```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# tool to convert 'melody' into 'data'
import pretty_midi
```

# Build data table

```python
# numerify the twelve notes of the western music theory
# A, Bb, C, C#, ..., G, G#
def principal_value(note):
    if note == 'A'                  : return 0
    if note == 'A#' or note == 'Bb' : return 1
    if note == 'B'                  : return 2
    if note == 'C'                  : return 3
    if note == 'C#' or note == 'Db' : return 4
    if note == 'D'                  : return 5
    if note == 'D#' or note == 'Eb' : return 6
    if note == 'E'                  : return 7
    if note == 'F'                  : return 8
    if note == 'F#' or note == 'Gb' : return 9
    if note == 'G'                  : return 10
    if note == 'G#' or note == 'Ab' : return 11

# numerify the note depending on octave number
# this corresponds to the key number on a 88-key piano, from A0 to C8
# will be used to get absolute distance between two notes
def value(note, octave):
    return octave * 12 + principal_value(note)

# numerify the note relative to the corresponding major scale of the piece
def canonical_value(note, key):
    return (principal_value(note) - principal_value(key)) % 12

# compute the corresponding duration for 60 BPM 
def canonical_duration(duration, BPM):
    return float(BPM) * duration / 60.0

quarter_note = 1.0
    
whole = 4.0 * quarter_note
half = 2.0 * quarter_note
quarter = 1.0 * quarter_note
eighth = 1.0 / 2.0 * quarter_note
sixteenth = 1.0 / 4.0 * quarter_note
thirtysecond = 1.0 / 8.0 * quarter_note
sixtyforth = 1.0 / 16.0 * quarter_note

half_dot = 1.5 * half
quarter_dot = 1.5 * quarter
eighth_dot = 1.5 * eighth
sixteenth_dot = 1.5 * sixteenth

half_3let = 2.0 / 3.0 * half
quarter_3let = 2.0 / 3.0 * quarter
eighth_3let = 2.0 / 3.0 * eighth
sixteenth_3let = 2.0 / 3.0 * sixteenth

#half_5let = 2.0 / 5.0 * half
#quarter_5let = 2.0 / 5.0 * quarter
#eighth_5let = 2.0 / 5.0 * eighth
#sixteenth_5let = 2.0 / 5.0 * sixteenth
#
#half_7let = 2.0 / 7.0 * half
#quarter_7let = 2.0 / 7.0 * quarter
#eighth_7let = 2.0 / 7.0 * eighth
#sixteenth_7let = 2.0 / 7.0 * sixteenth

# i'll ignore anything else
# these are sufficient enough for covering the data and for creativity, as well

# sorted for comparison purposes 
#durations = np.array([whole, half_dot, half, quarter_dot, half_3let, quarter, half_5let,
#                      eighth_dot, quarter_3let, half_7let, eighth, quarter_5let,
#                      sixteenth_dot, eighth_3let, quarter_7let, sixteenth, eighth_5let,
#                      sixteenth_3let, eighth_7let, thirtysecond, sixteenth_5let, sixteenth_7let])
#
#labels = ['whole', 'half_dot', 'half', 'quarter_dot', 'half_3let', 'quarter', 'half_5let',
#          'eighth_dot', 'quarter_3let', 'half_7let', 'eighth', 'quarter_5let',
#          'sixteenth_dot', 'eighth_3let', 'quarter_7let', 'sixteenth', 'eighth_5let',
#          'sixteenth_3let', 'eighth_7let', 'thirtysecond', 'sixteenth_5let', 'sixteenth_7let']
durations = np.array([whole, half_dot, half, quarter_dot, half_3let, quarter,
                      eighth_dot, quarter_3let, eighth,
                      sixteenth_dot, eighth_3let, sixteenth,
                      sixteenth_3let, thirtysecond])

labels = ['whole', 'half_dot', 'half', 'quarter_dot', 'half_3let', 'quarter',
          'eighth_dot', 'quarter_3let', 'eighth',
          'sixteenth_dot', 'eighth_3let', 'sixteenth',
          'sixteenth_3let', 'thirtysecond']

def canonical_duration_label(duration):
    return np.argmin(np.abs(durations - duration))

# nocturne numbers
pieces = [1, 2, 6, 11, 20, 21]
pieces2 = [1, 2, 6, 11, 20, 21, 0]

# nocturne keys
keys = ['C#', 'D#', 'A#', 'A#', 'A', 'D#']
keys2 = ['C#', 'D#', 'A#', 'A#', 'A', 'D#', 'G']

# nocturne BPMs
# indicated BPM times correction factor to fix mismatches in MIDI encoding
BPMs = [90.0 * 0.5 / 0.472916903125, 
        60.0 * 0.5 / 0.47291666666666665, 
        100.0 * 1.0 / 0.9479166666666666, 
        66.0 * 2.0 / 1.9123992891666668, 
        60.0 * 1.0 / 0.9479166666666666, 
        60.0 * 1.0 / 0.9979166666666667]
BPMs2 = [90.0 * 0.5 / 0.472916903125, 
         60.0 * 0.5 / 0.47291666666666665, 
         100.0 * 1.0 / 0.9479166666666666, 
         66.0 * 2.0 / 1.9123992891666668, 
         60.0 * 1.0 / 0.9479166666666666, 
         60.0 * 1.0 / 0.9979166666666667,
         80.0 * 1.0 / 0.9979166666666668]

# midi files
midis = ['./src/midi/' + str(i) + '.mid' for i in pieces]
midis2 = ['./src/midi/' + str(i) + '.mid' for i in pieces2]

# data files
raw_data_file = './src/data/raw/combined_data.csv'
raw_data_file2 = './src/data/raw/combined_data2.csv'

# export data table for piece i = range(len(pieces)), not i = piece
def prepare_data_table(i):
    midi = midis[i]
    midi_data = pretty_midi.PrettyMIDI(midi)

    # raw data table
    # 0:note 
    # 1:octave 
    # 2:duration (s)
    raw = [(note_name[:-1], int(note_name[-1]), note.end - note.start)
           for note in midi_data.instruments[0].notes
           if (note_name := pretty_midi.note_number_to_name(note.pitch))]
    
    # melody data table
    # 0:note
    # 1:principal value
    # 2:value
    # 3:canonical value
    # 4:octave
    # 5:duration (s)
    # 6:canonical duration (s)
    # 7:canonical duration label
    # 8:weight: we'll explain this later
    melody = [] 
    for row in raw:
        note = row[0]
        octave = row[1]
        duration = row[2]
        key = keys[i]
        BPM = BPMs[i]
        
        pv = principal_value(note)
        val = value(note, octave)
        cval = canonical_value(note, key)
        dur = duration
        cdur = canonical_duration(duration, BPM)
        cdurl = canonical_duration_label(cdur)
        weight = 1.0
        
        melody.append((note, pv, val, cval, octave, dur, cdur, cdurl, weight))

    return melody

with open(raw_data_file, 'w') as file:
    head = ('0:note', '1:principal value', '2:value', '3:canonical value', 
            '4:octave', '5:duration', '6:canonical duration', 
            '7:canonical duration label', '8:weight')
    file.write(','.join(map(str, head)) + "\n")
    
    for i in range(len(pieces)):
        for row in prepare_data_table(i):
            file.write(','.join(map(str, row)) + "\n")

# export data table for piece i = range(len(pieces)), not i = piece
def prepare_data_table2(i):
    midi = midis2[i]
    midi_data = pretty_midi.PrettyMIDI(midi)

    # raw data table
    # 0:note 
    # 1:octave 
    # 2:duration (s)
    raw = [(note_name[:-1], int(note_name[-1]), note.end - note.start)
           for note in midi_data.instruments[0].notes
           if (note_name := pretty_midi.note_number_to_name(note.pitch))]
    
    # melody data table
    # 0:note
    # 1:principal value
    # 2:value
    # 3:canonical value
    # 4:octave
    # 5:duration (s)
    # 6:canonical duration (s)
    # 7:canonical duration label
    # 8:weight: the 0th piece is my sample melody for the computer to complete
    #           i'll introduce them with a weight of 10 to emphasize my style
    melody = [] 
    for row in raw:
        note = row[0]
        octave = row[1]
        duration = row[2]
        key = keys2[i]
        BPM = BPMs2[i]
        
        pv = principal_value(note)
        val = value(note, octave)
        cval = canonical_value(note, key)
        dur = duration
        cdur = canonical_duration(duration, BPM)
        cdurl = canonical_duration_label(cdur)
        weight = 10.0 if i == 6 else 1.0
        
        melody.append((note, pv, val, cval, octave, dur, cdur, cdurl, weight))

    return melody

with open(raw_data_file2, 'w') as file:
    head = ('0:note', '1:principal value', '2:value', '3:canonical value', 
            '4:octave', '5:duration', '6:canonical duration', 
            '7:canonical duration label', '8:weight')
    file.write(','.join(map(str, head)) + "\n")
    
    for i in range(len(pieces2)):
        for row in prepare_data_table2(i):
            file.write(','.join(map(str, row)) + "\n")

# tools for engineered features
def are_consonant(note_1, note_2):
    # assumes canonical values
    lower = np.min([note_1, note_2])
    higher = np.max([note_1, note_2])
    
    unison = higher - lower == 0
    #second = higher - lower == 2
    minor_third = higher - lower == 3
    major_third = higher - lower == 4
    #fourth = higher - lower == 5
    fifth = higher - lower == 7
    minor_seventh = higher - lower == 10
    major_seventh = higher - lower == 11

    #consonant = unison or second or minor_third or major_third or fourth or fifth or minor_seventh or major_seventh
    consonant = unison or minor_third or major_third or fifth or minor_seventh or major_seventh
    
    return 1 if consonant else 0

def is_in_key(note):
    # assumes canonical note
    return 1 if note in [0, 2, 4, 5, 7, 9, 11] else 0

# grand data table with a grouping parameter g
# where a group G is g consecutive notes 
# G = {note_1, note_2, ..., note_g}
# we consider g = 3, 4, 5

# import data back
raw_dt = pd.read_csv(raw_data_file)

vals = np.array(raw_dt.T.iloc[2].values, int)
cvals = np.array(raw_dt.T.iloc[3].values, int)
octaves = np.array(raw_dt.T.iloc[4].values, int)
cdurls = np.array(raw_dt.T.iloc[7].values, str)

##########################
# grouping parameter = 3 #
##########################

dt = [('0:index', '1:weight', 
       
       # targets
       '2:y1 (can val)', 
       '3:y2 (octave)', 
       '4:y3 (can dur lbl)',
       
       # natural features
       '5:x11 (can val.)', 
       '6:x12 (octave)', 
       '7:x13 (can dur lbl)',
       
       '8:x21 (can val)', 
       '9:x22 (octave)', 
       '10:x23 (can dur lbl)',
       
       # engineered features
       '11:x1 in key',
       '12:x2 in key',
       '13:x2 - x1 (diff val)',
       '14:x1 x2 consonant',
       '15:#occur of x1 so far',
       '16:#occur of x2 so far',
       '17:mov avg of all prev. notes incl. x1 x2',
       '18:x2 < x1 dur')]

for p in range(len(raw_dt) - 2):
    index = p
    weight = 1.0

    # targets    
    y1 = cvals[p + 2]
    y2 = octaves[p + 2]
    y3 = cdurls[p + 2]
    
    # natural features
    x11 = cvals[p]
    x12 = octaves[p]
    x13 = cdurls[p]
    
    x21 = cvals[p + 1]
    x22 = octaves[p + 1]
    x23 = cdurls[p + 1]
    
    # engineered features
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    dx21 = vals[p + 1] - vals[p]
    cx21 = are_consonant(x11, x21)
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    mov_avg = float(np.sum(cvals[0:p + 1])) / float(p + 3)
    dur21 = 1.0 if x23 > x13 else 0.0
    
    dt.append((index, weight, 
               y1, y2, y3,
               x11, x12, x13, 
               x21, x22, x23, 
               x1_in_key, x2_in_key, 
               dx21, cx21, 
               occur_x1, occur_x2, 
               mov_avg,
               dur21))

data_file = './src/data/gp3.csv'
with open(data_file, 'w') as file:
    for row in dt:
        file.write(','.join(map(str, row)) + "\n")

##########################
# grouping parameter = 4 #
##########################

dt = [('0:index', '1:weight', 
       
       # targets
       '2:y1 (can val)', 
       '3:y2 (octave)', 
       '4:y3 (can dur lbl)',
       
       # natural features
       '5:x11 (can val.)', 
       '6:x12 (octave)', 
       '7:x13 (can dur lbl)',
       
       '8:x21 (can val)', 
       '9:x22 (octave)', 
       '10:x23 (can dur lbl)',

       '11:x31 (can val)', 
       '12:x32 (octave)', 
       '13:x33 (can dur lbl)',
       
       # engineered features
       '14:x1 in key',
       '15:x2 in key',
       '16:x3 in key',
       '17:x2 - x1 (diff val)',
       '18:x3 - x1 (diff val)',
       '19:x3 - x2 (diff val)',
       '20:x1 x2 consonant',
       '21:x1 x3 consonant',
       '22:x2 x3 consonant',
       '23:#occur of x1 so far',
       '24:#occur of x2 so far',
       '25:#occur of x3 so far',
       '26:mov avg of all prev. notes incl. x1 x2 x3',
       '27:x2 < x1 dur',
       '28:x3 < x1 dur',
       '29:x3 < x2 dur')]

for p in range(len(raw_dt) - 3):
    index = p
    weight = 1.0

    # targets    
    y1 = cvals[p + 3]
    y2 = octaves[p + 3]
    y3 = cdurls[p + 3]
    
    # natural features
    x11 = cvals[p]
    x12 = octaves[p]
    x13 = cdurls[p]
    
    x21 = cvals[p + 1]
    x22 = octaves[p + 1]
    x23 = cdurls[p + 1]

    x31 = cvals[p + 2]
    x32 = octaves[p + 2]
    x33 = cdurls[p + 2]
    
    # engineered features
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx32 = are_consonant(x21, x31)
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    mov_avg = float(np.sum(cvals[0:p + 2])) / float(p + 3)
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0
    
    dt.append((index, weight, 
               y1, y2, y3,
               x11, x12, x13, 
               x21, x22, x23, 
               x31, x32, x33,
               x1_in_key, x2_in_key, x3_in_key, 
               dx21, dx31, dx32,
               cx21, cx31, cx32, 
               occur_x1, occur_x2, occur_x3, 
               mov_avg,
               dur21, dur31, dur32))

data_file = './src/data/gp4.csv'
with open(data_file, 'w') as file:
    for row in dt:
        file.write(','.join(map(str, row)) + "\n")

##########################
# grouping parameter = 5 #
##########################

dt = [('0:index', '1:weight', 
       
       # targets
       '2:y1 (can val)', 
       '3:y2 (octave)', 
       '4:y3 (can dur lbl)',
       
       # natural features
       '5:x11 (can val.)', 
       '6:x12 (octave)', 
       '7:x13 (can dur lbl)',
       
       '8:x21 (can val)', 
       '9:x22 (octave)', 
       '10:x23 (can dur lbl)',

       '11:x31 (can val)', 
       '12:x32 (octave)', 
       '13:x33 (can dur lbl)',

       '14:x41 (can val)', 
       '15:x42 (octave)', 
       '16:x43 (can dur lbl)',
       
       # engineered features
       '17:x1 in key',
       '18:x2 in key',
       '19:x3 in key',
       '20:x4 in key',
       '21:x2 - x1 (diff val)',
       '22:x3 - x1 (diff val)',
       '23:x4 - x1 (diff val)',
       '24:x3 - x2 (diff val)',
       '25:x4 - x2 (diff val)',
       '26:x4 - x3 (diff val)',
       '27:x1 x2 consonant',
       '28:x1 x3 consonant',
       '29:x1 x4 consonant',
       '30:x2 x3 consonant',
       '31:x2 x4 consonant',
       '32:x3 x4 consonant',
       '33:#occur of x1 so far',
       '34:#occur of x2 so far',
       '35:#occur of x3 so far',
       '36:#occur of x4 so far',
       '37:mov avg of all prev. notes incl. x1 x2 x3 x4',
       '38:x2 < x1 dur',
       '39:x3 < x1 dur',
       '40:x4 < x1 dur',
       '41:x3 < x2 dur',
       '42:x4 < x2 dur',
       '43:x4 < x3 dur')]

for p in range(len(raw_dt) - 4):
    index = p
    weight = 1.0

    # targets    
    y1 = cvals[p + 4]
    y2 = octaves[p + 4]
    y3 = cdurls[p + 4]
    
    # natural features
    x11 = cvals[p]
    x12 = octaves[p]
    x13 = cdurls[p]
    
    x21 = cvals[p + 1]
    x22 = octaves[p + 1]
    x23 = cdurls[p + 1]

    x31 = cvals[p + 2]
    x32 = octaves[p + 2]
    x33 = cdurls[p + 2]

    x41 = cvals[p + 3]
    x42 = octaves[p + 3]
    x43 = cdurls[p + 3]
    
    # engineered features
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    x4_in_key = is_in_key(x41)
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx41 = vals[p + 3] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    dx42 = vals[p + 3] - vals[p + 1]
    dx43 = vals[p + 3] - vals[p + 2]
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx41 = are_consonant(x11, x41)
    cx32 = are_consonant(x21, x31)
    cx42 = are_consonant(x21, x41)
    cx43 = are_consonant(x31, x41)
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    occur_x4 = np.sum(cvals[0:p] == x41)
    mov_avg = float(np.sum(cvals[0:p + 3])) / float(p + 4)
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur41 = 1.0 if x43 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0
    dur42 = 1.0 if x43 > x23 else 0.0
    dur43 = 1.0 if x43 > x33 else 0.0
    
    dt.append((index, weight, 
               y1, y2, y3,
               x11, x12, x13, 
               x21, x22, x23, 
               x31, x32, x33,
               x41, x42, x43,
               x1_in_key, x2_in_key, x3_in_key, x4_in_key,
               dx21, dx31, dx41, dx32, dx42, dx43,
               cx21, cx31, cx41, cx32, cx42, cx43, 
               occur_x1, occur_x2, occur_x3, occur_x4,
               mov_avg,
               dur21, dur31, dur41, dur32, dur42, dur43))

data_file = './src/data/gp5.csv'
with open(data_file, 'w') as file:
    for row in dt:
        file.write(','.join(map(str, row)) + "\n")

##########################
# grouping parameter = 6 #
##########################

dt = [('0:index', '1:weight', 
       
       # targets
       '2:y1 (can val)', 
       '3:y2 (octave)', 
       '4:y3 (can dur lbl)',
       
       # natural features
       '5:x11 (can val.)', 
       '6:x12 (octave)', 
       '7:x13 (can dur lbl)',
       
       '8:x21 (can val)', 
       '9:x22 (octave)', 
       '10:x23 (can dur lbl)',

       '11:x31 (can val)', 
       '12:x32 (octave)', 
       '13:x33 (can dur lbl)',

       '14:x41 (can val)', 
       '15:x42 (octave)', 
       '16:x43 (can dur lbl)',

       '17:x51 (can val)', 
       '18:x52 (octave)', 
       '19:x53 (can dur lbl)',
       
       # engineered features
       '20:x1 in key',
       '21:x2 in key',
       '22:x3 in key',
       '23:x4 in key',
       '24:x5 in key',
       '25:x2 - x1 (diff val)',
       '26:x3 - x1 (diff val)',
       '27:x4 - x1 (diff val)',
       '28:x5 - x1 (diff val)',
       '29:x3 - x2 (diff val)',
       '30:x4 - x2 (diff val)',
       '31:x5 - x2 (diff val)',
       '32:x4 - x3 (diff val)',
       '33:x5 - x3 (diff val)',
       '34:x5 - x4 (diff val)',
       '35:x1 x2 consonant',
       '36:x1 x3 consonant',
       '37:x1 x4 consonant',
       '38:x1 x5 consonant',
       '39:x2 x3 consonant',
       '40:x2 x4 consonant',
       '41:x2 x5 consonant',
       '42:x3 x4 consonant',
       '43:x3 x5 consonant',
       '44:x4 x5 consonant',
       '45:#occur of x1 so far',
       '46:#occur of x2 so far',
       '47:#occur of x3 so far',
       '48:#occur of x4 so far',
       '49:#occur of x5 so far',
       '50:mov avg of all prev. notes incl. x1 x2 x3 x4 x5',
       '51:x2 < x1 dur',
       '52:x3 < x1 dur',
       '53:x4 < x1 dur',
       '54:x5 < x1 dur',
       '55:x3 < x2 dur',
       '56:x4 < x2 dur',
       '57:x5 < x2 dur',
       '58:x4 < x3 dur',
       '59:x5 < x3 dur',
       '60:x5 < x4 dur')]

for p in range(len(raw_dt) - 5):
    index = p
    weight = 1.0

    # targets    
    y1 = cvals[p + 5]
    y2 = octaves[p + 5]
    y3 = cdurls[p + 5]
    
    # natural features
    x11 = cvals[p]
    x12 = octaves[p]
    x13 = cdurls[p]
    
    x21 = cvals[p + 1]
    x22 = octaves[p + 1]
    x23 = cdurls[p + 1]

    x31 = cvals[p + 2]
    x32 = octaves[p + 2]
    x33 = cdurls[p + 2]

    x41 = cvals[p + 3]
    x42 = octaves[p + 3]
    x43 = cdurls[p + 3]

    x51 = cvals[p + 4]
    x52 = octaves[p + 4]
    x53 = cdurls[p + 4]
    
    # engineered features
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    x4_in_key = is_in_key(x41)
    x5_in_key = is_in_key(x51)
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx41 = vals[p + 3] - vals[p]
    dx51 = vals[p + 4] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    dx42 = vals[p + 3] - vals[p + 1]
    dx52 = vals[p + 4] - vals[p + 1]
    dx43 = vals[p + 3] - vals[p + 2]
    dx53 = vals[p + 4] - vals[p + 2]
    dx54 = vals[p + 4] - vals[p + 3]
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx41 = are_consonant(x11, x41)
    cx51 = are_consonant(x11, x51)
    cx32 = are_consonant(x21, x31)
    cx42 = are_consonant(x21, x41)
    cx52 = are_consonant(x21, x51)
    cx43 = are_consonant(x31, x41)
    cx53 = are_consonant(x31, x51)
    cx54 = are_consonant(x41, x51)
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    occur_x4 = np.sum(cvals[0:p] == x41)
    occur_x5 = np.sum(cvals[0:p] == x51)
    mov_avg = float(np.sum(cvals[0:p + 4])) / float(p + 5)
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur41 = 1.0 if x43 > x13 else 0.0
    dur51 = 1.0 if x53 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0
    dur42 = 1.0 if x43 > x23 else 0.0
    dur52 = 1.0 if x53 > x23 else 0.0
    dur43 = 1.0 if x43 > x33 else 0.0
    dur53 = 1.0 if x53 > x33 else 0.0
    dur54 = 1.0 if x53 > x43 else 0.0
    
    dt.append((index, weight, 
               y1, y2, y3,
               x11, x12, x13, 
               x21, x22, x23, 
               x31, x32, x33,
               x41, x42, x43,
               x51, x52, x53,
               x1_in_key, x2_in_key, x3_in_key, x4_in_key, x5_in_key,
               dx21, dx31, dx41, dx51, dx32, dx42, dx52, dx43, dx53, dx54,
               cx21, cx31, cx41, cx51, cx32, cx42, cx52, cx43, cx53, cx54, 
               occur_x1, occur_x2, occur_x3, occur_x4, occur_x5,
               mov_avg,
               dur21, dur31, dur41, dur51, dur32, dur42, dur52, dur43, dur53, dur54))

data_file = './src/data/gp6.csv'
with open(data_file, 'w') as file:
    for row in dt:
        file.write(','.join(map(str, row)) + "\n")
```

```python
##########################
# grouping parameter = 3 #
##########################

dt = pd.read_csv('./src/data/gp3.csv')

P = len(dt)

# we have 3 targets
# theoretical values for the number of classes:
# canonical value: 0, 1, 2, ..., 11 => C_1 = 12
# octave: 0, 1, 2, ..., 8 => C_2 = 9
# canonical duration label: len(labels)
y1 = np.array(dt.T.iloc[2].values, int)
y2 = np.array(dt.T.iloc[3].values, int)
y3 = np.array(dt.T.iloc[4].values, int)

C1 = 12
C2 = 9
C3 = len(labels)

# we have 6 natural, 8 engineered features
N = 14

x1 = np.array(dt.T.iloc[5].values, float) 
x2 = np.array(dt.T.iloc[6].values, float) 
x3 = np.array(dt.T.iloc[7].values, float) 
x4 = np.array(dt.T.iloc[8].values, float) 
x5 = np.array(dt.T.iloc[9].values, float) 
x6 = np.array(dt.T.iloc[10].values, float)

x7 = np.array(dt.T.iloc[11].values, float) 
x8 = np.array(dt.T.iloc[12].values, float)
x9 = np.array(dt.T.iloc[13].values, float)
x10 = np.array(dt.T.iloc[14].values, float)
x11 = np.array(dt.T.iloc[15].values, float)
x12 = np.array(dt.T.iloc[16].values, float)
x13 = np.array(dt.T.iloc[17].values, float)
x14 = np.array(dt.T.iloc[18].values, float)

# append 1 for bias 
xo = np.column_stack((np.ones(P), x1, x2, x3, x4, x5, x6, x7, x8, x9, 
                      x10, x11, x12, x13, x14))

# multiclass softmax
def g(w, y):
    dots = np.dot(xo, w.T)

    max_dots = np.max(dots, axis = 1, keepdims = True)
    stabilized_dots = dots - max_dots

    exp_dots = np.exp(stabilized_dots)
    softmax_den = np.sum(exp_dots, axis = 1, keepdims = True)

    correct_class_dots = stabilized_dots[np.arange(P), y]
    log_probs = correct_class_dots - np.log(softmax_den.flatten())

    return -np.mean(log_probs)

def dg(w, y):
    dots = np.dot(xo, w.T)

    max_dots = np.max(dots, axis = 1, keepdims = True)
    stabilized_dots = dots - max_dots

    exp_dots = np.exp(stabilized_dots)
    softmax_den = np.sum(exp_dots, axis = 1, keepdims = True)
    probs = exp_dots / softmax_den

    one_hot = np.zeros_like(probs)
    one_hot[np.arange(P), y] = 1

    grad = -np.dot((one_hot - probs).T, xo) / P

    return grad

K = 2000

w1 = 0.005 * np.ones((C1, N + 1))
w2 = 0.005 * np.ones((C2, N + 1))
w3 = 0.005 * np.ones((C3, N + 1))

hist1 = []
hist2 = []
hist3 = []

for k in range(K):
    alpha1 = 0.1 / float(k + 1)
    alpha2 = 0.1 / float(k + 1)
    alpha3 = 0.01
    w1 = w1 - alpha1 * dg(w1, y1)
    w2 = w2 - alpha2 * dg(w2, y2)
    w3 = w3 - alpha3 * dg(w3, y3)
    hist1.append(g(w1, y1)) 
    hist2.append(g(w2, y2)) 
    hist3.append(g(w3, y3)) 
    print(f"\rprogress: {k + 1} / {K}", end = '', flush = True)

predic1 = np.array([np.argmax([np.dot(xo[p], w1[c]) for c in range(C1)]) for p in range(P)])
percent_mismatch1 = round(np.sum(y1 != predic1) * 100.0 / len(y1))
predic2 = np.array([np.argmax([np.dot(xo[p], w2[c]) for c in range(C2)]) for p in range(P)])
percent_mismatch2 = round(np.sum(y2 != predic2) * 100.0 / len(y2))
predic3 = np.array([np.argmax([np.dot(xo[p], w3[c]) for c in range(C3)]) for p in range(P)])
percent_mismatch3 = round(np.sum(y3 != predic3) * 100.0 / len(y3))

print('\n')
print('% mismatch 1:', percent_mismatch1)
print('% mismatch 2:', percent_mismatch2)
print('% mismatch 3:', percent_mismatch3)

plt.plot(range(1, K+1), hist1, label = '1')
plt.plot(range(1, K+1), hist2, label = '2')
plt.plot(range(1, K+1), hist3, label = '3')
plt.xscale('log')
plt.legend()
plt.show()

w_3 = [w1, w2, w3]
```

```python
def get_features_3(x1, x2):
    x11, x12, x13 = x1
    x21, x22, x23 = x2

    p = 0
        
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    
    dx21 = vals[p + 1] - vals[p]
    
    cx21 = are_consonant(x11, x21)
    
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    
    mov_avg = float(np.sum(cvals[0:p + 1])) / float(p + 2)
    
    dur21 = 1.0 if x23 > x13 else 0.0

    x = np.array([1.0, x11, x12, x13, x21, x22, x23, 
                  x1_in_key, x2_in_key,  
                  dx21, 
                  cx21, 
                  occur_x1, occur_x2, 
                  mov_avg,
                  dur21])
    
    return x

def get_features_4(x1, x2, x3):
    x11, x12, x13 = x1
    x21, x22, x23 = x2
    x31, x32, x33 = x3

    p = 0
        
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx32 = are_consonant(x21, x31)
    
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    
    mov_avg = float(np.sum(cvals[0:p + 2])) / float(p + 3)
    
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0    

    x = np.array([1.0, x11, x12, x13, x21, x22, x23, x31, x32, x33, 
                  x1_in_key, x2_in_key, x3_in_key, 
                  dx21, dx31, dx32, 
                  cx21, cx31, cx32, 
                  occur_x1, occur_x2, occur_x3, 
                  mov_avg,
                  dur21, dur31, dur32])
    
    return x

def get_features_5(x1, x2, x3, x4):
    x11, x12, x13 = x1
    x21, x22, x23 = x2
    x31, x32, x33 = x3
    x41, x42, x43 = x4

    p = 0
        
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    x4_in_key = is_in_key(x41)
    
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx41 = vals[p + 3] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    dx42 = vals[p + 3] - vals[p + 1]
    dx43 = vals[p + 3] - vals[p + 2]
    
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx41 = are_consonant(x11, x41)
    cx32 = are_consonant(x21, x31)
    cx42 = are_consonant(x21, x41)
    cx43 = are_consonant(x31, x41)
    
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    occur_x4 = np.sum(cvals[0:p] == x41)
    
    mov_avg = float(np.sum(cvals[0:p + 3])) / float(p + 4)
    
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur41 = 1.0 if x43 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0
    dur42 = 1.0 if x43 > x23 else 0.0
    dur43 = 1.0 if x43 > x33 else 0.0    

    x = np.array([1.0, x11, x12, x13, x21, x22, x23, x31, x32, x33, x41, x42, x43, 
                  x1_in_key, x2_in_key, x3_in_key, x4_in_key, 
                  dx21, dx31, dx41, dx32, dx42, dx43,
                  cx21, cx31, cx41, cx32, cx42, cx43, 
                  occur_x1, occur_x2, occur_x3, occur_x4, 
                  mov_avg,
                  dur21, dur31, dur41, dur32, dur42, dur43])
    
    return x

def get_features_6(x1, x2, x3, x4, x5):
    x11, x12, x13 = x1
    x21, x22, x23 = x2
    x31, x32, x33 = x3
    x41, x42, x43 = x4
    x51, x52, x53 = x5

    p = 0
        
    x1_in_key = is_in_key(x11)
    x2_in_key = is_in_key(x21)
    x3_in_key = is_in_key(x31)
    x4_in_key = is_in_key(x41)
    x5_in_key = is_in_key(x51)
    
    dx21 = vals[p + 1] - vals[p]
    dx31 = vals[p + 2] - vals[p]
    dx41 = vals[p + 3] - vals[p]
    dx51 = vals[p + 4] - vals[p]
    dx32 = vals[p + 2] - vals[p + 1]
    dx42 = vals[p + 3] - vals[p + 1]
    dx52 = vals[p + 4] - vals[p + 1]
    dx43 = vals[p + 3] - vals[p + 2]
    dx53 = vals[p + 4] - vals[p + 2]
    dx54 = vals[p + 4] - vals[p + 3]
    
    cx21 = are_consonant(x11, x21)
    cx31 = are_consonant(x11, x31)
    cx41 = are_consonant(x11, x41)
    cx51 = are_consonant(x11, x51)
    cx32 = are_consonant(x21, x31)
    cx42 = are_consonant(x21, x41)
    cx52 = are_consonant(x21, x51)
    cx43 = are_consonant(x31, x41)
    cx53 = are_consonant(x31, x51)
    cx54 = are_consonant(x41, x51)
    
    occur_x1 = np.sum(cvals[0:p] == x11)
    occur_x2 = np.sum(cvals[0:p] == x21)
    occur_x3 = np.sum(cvals[0:p] == x31)
    occur_x4 = np.sum(cvals[0:p] == x41)
    occur_x5 = np.sum(cvals[0:p] == x51)
    
    mov_avg = float(np.sum(cvals[0:p + 4])) / float(p + 5)
    
    dur21 = 1.0 if x23 > x13 else 0.0
    dur31 = 1.0 if x33 > x13 else 0.0
    dur41 = 1.0 if x43 > x13 else 0.0
    dur51 = 1.0 if x53 > x13 else 0.0
    dur32 = 1.0 if x33 > x23 else 0.0
    dur42 = 1.0 if x43 > x23 else 0.0
    dur52 = 1.0 if x53 > x23 else 0.0
    dur43 = 1.0 if x43 > x33 else 0.0
    dur53 = 1.0 if x53 > x33 else 0.0
    dur54 = 1.0 if x53 > x43 else 0.0    

    x = np.array([1.0, x11, x12, x13, x21, x22, x23, x31, x32, x33, x41, x42, x43, x51, x52, x53,
                  x1_in_key, x2_in_key, x3_in_key, x4_in_key, x5_in_key,
                  dx21, dx31, dx41, dx51, dx32, dx42, dx52, dx43, dx53, dx54,
                  cx21, cx31, cx41, cx51, cx32, cx42, cx52, cx43, cx53, cx54, 
                  occur_x1, occur_x2, occur_x3, occur_x4, occur_x5,
                  mov_avg,
                  dur21, dur31, dur41, dur51, dur32, dur42, dur52, dur43, dur53, dur54])
    
    return x

BPM = 60
time_signature = (4, 4)
key = 'F' # major or D minor
key_value = principal_value(key)
target_duration = 60 # seconds

def notify(n):
    n1, n2, n3 = n
    n1 = (n1 + key_value) % 12
    if n1 == 0:
        label = 'A'
    elif n1 == 1:
        label = 'A#'
    elif n1 == 2:
        label = 'B'
    elif n1 == 3:
        label = 'C'
    elif n1 == 4:
        label = 'C#'
    elif n1 == 5:
        label = 'D'
    elif n1 == 6:
        label = 'D#'
    elif n1 == 7:
        label = 'E'
    elif n1 == 8:
        label = 'F'
    elif n1 == 9:
        label = 'F#'
    elif n1 == 10:
        label = 'G'
    elif n1 == 11:
        label = 'G#'
    dur = durations[n3] * 60.0 / float(BPM)
    return [label + str(n2), dur]

# an explicit example with grouping parameter = 6
compose = []

n1 = [cvals_first_6[0], octaves_first_6[0], cdurls_first_6[0]]
n2 = [cvals_first_6[1], octaves_first_6[1], cdurls_first_6[1]]
n3 = [cvals_first_6[2], octaves_first_6[2], cdurls_first_6[2]]
n4 = [cvals_first_6[3], octaves_first_6[3], cdurls_first_6[3]]
n5 = [cvals_first_6[4], octaves_first_6[4], cdurls_first_6[4]]

compose.append(notify(n1))
compose.append(notify(n2))
compose.append(notify(n3))
compose.append(notify(n4))
compose.append(notify(n5))

elapsed_time = 0.0

while elapsed_time < target_duration:
    n6_cval_probs = np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[0].T)) / np.sum(np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[0].T)))
    n6_octave_probs = np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[1].T)) / np.sum(np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[1].T)))
    n6_cdurl_probs = np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[2].T)) / np.sum(np.exp(np.dot(get_features_6(n1, n2, n3, n4, n5), w_6[2].T)))
    n6 = [np.random.choice(len(n6_cval_probs), p = n6_cval_probs), 
          np.random.choice(len(n6_octave_probs), p = n6_octave_probs), 
          np.random.choice(len(n6_cdurl_probs), p = n6_cdurl_probs)]

    compose.append(notify(n6))

    elapsed_time = np.sum(np.array(np.array(compose)[:, 1], float))

    n1 = n2
    n2 = n3
    n3 = n4
    n4 = n5
    n5 = n6

def note_to_midi(note):
    note_names = {'C': 0, 'C#': 1, 'D': 2, 'D#': 3, 'E': 4, 'F': 5, 'F#': 6,
                  'G': 7, 'G#': 8, 'A': 9, 'A#': 10, 'B': 11}
    octave = int(note[-1])
    note_base = note[:-1]
    return 12 * (octave + 1) + note_names[note_base]

def generate_midi(note_list, bpm, ts, output_file='./out/composition.mid'):
    midi = pretty_midi.PrettyMIDI()
    tempo = bpm
    midi._tempo_changes = (0.0, [tempo])
    midi.time_signature_changes.append(pretty_midi.TimeSignature(numerator=ts[0], denominator=ts[1], time=0.0))
    instrument = pretty_midi.Instrument(program=pretty_midi.instrument_name_to_program('Acoustic Grand Piano'))
    seconds_per_beat = 60.0 / bpm
    current_time = 0.0

    for note, duration in note_list:
        midi_note = note_to_midi(note)
        note_duration = duration  
        instrument.notes.append(pretty_midi.Note(
            velocity = 100,  
            pitch = midi_note,  
            start = current_time, 
            end = current_time + note_duration 
        ))

        current_time += note_duration

    midi.instruments.append(instrument)
    midi.write(output_file)

generate_midi(compose, BPM, time_signature)
```
